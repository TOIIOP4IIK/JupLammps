```python
#Импорт необходимых библиотек
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
Metal = "Ti"
Sigma = "150.3"
IDDel = 30
print(Metal)
print(Sigma)
```

```python
#Получить данные по размеру блока!!start
BoxSim = pd.read_csv('STR/'+ Metal + '_'+Sigma+'.lmp' , header = None, skiprows = 5, nrows = 3, sep = ' ' )
BoxSim
```

```python
Xbeg = BoxSim.iloc[0][0]
Xend = BoxSim.iloc[0][1]
Ybeg = BoxSim.iloc[1][0]
Yend = BoxSim.iloc[1][1]
Zbeg = BoxSim.iloc[2][0]
Zend = BoxSim.iloc[2][1]
X = Xend - Xbeg
Y = Yend - Ybeg
Z = Zend - Zbeg
print("Размер блока по Х = ",X)
print("Размер блока по Y = ",Y)
print("Размер блока по Z = ",Z)
```

```python
#Чтение файла блока и получение данных в таблицу
AtomId = pd.read_csv('STR/'+Metal+'_'+Sigma+'.lmp' , header = 9, sep = ' ' , names = ["x","y","z","Type","id"])
AtomId
```

```python
#ДатаФрейм !!!!!
#выводит инфу по строкам
#df.columns
#Фильтр для значений таблицы
filter_x = AtomId['x'] > ((X/2)-6) 
filter_x2 = AtomId['x'] < ((X/2)+6) 
filter_y = AtomId['y'] > (Y/4)
filter_y2 = AtomId['y'] < (Y/4)*3
filter_z = AtomId['z'] > (Z/2)-3
filter_z2 = AtomId['z'] < (Z/2)+3
a = AtomId.loc[filter_x & filter_x2 & filter_y & filter_y2 & filter_z & filter_z2]
```

```python
#Подсчет количества отфильтрованных строк!!
len(a)
```

```python
#Вывод отфильтрованной информации в виде ДатаФрейма
Da = pd.DataFrame(a)
print(Da)
```

```python
#Получение id атомов из файла
for i in range(len(a)):
    print(Da.iloc[i]['id'])
```
